Below I tried to mask these word to an image of Dog and Cat with the name used for them respectively.

```python
custom_mask_dog = np.array(Image.open('dog.png'))

def transform_format_dog(val):
    if val == 0:
        return 255
    else:
        return val

transformed_mask_dog = np.ndarray((custom_mask_dog.shape[0],custom_mask_dog.shape[1]), np.int32)

for i in range(len(transformed_mask_dog)):
    transformed_mask_dog[i] = list(map(transform_format_dog, custom_mask_dog[i]))


fig, ax = plt.subplots(figsize = (16, 12))
plt.subplot(1, 2, 1)

dogname = ' '.join(train.loc[train['Type'] == 1, 'Name'].fillna('').values)
wordcloud = WordCloud(stopwords=STOPWORDS, 
               background_color="white",
               max_words=300, max_font_size=256, width=custom_mask_dog.shape[1],
               height=custom_mask_dog.shape[0], mask=transformed_mask_dog, contour_width=3, contour_color='steelblue')
wordcloud.generate(dogname)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top Dog names')
```

```python
custom_mask_cat = np.array(Image.open('cat.png'))

def transform_format_cat(val):
    if val == 0:
        return 255
    else:
        return val

transformed_mask_cat = np.ndarray((custom_mask_cat.shape[0],custom_mask_cat.shape[1]), np.int32)

for i in range(len(transformed_mask_cat)):
    transformed_mask_cat[i] = list(map(transform_format_cat, custom_mask_cat[i]))


fig, ax = plt.subplots(figsize = (16, 12))
plt.subplot(1, 2, 1)

catname = ' '.join(train.loc[train['Type'] == 2, 'Name'].fillna('').values)
wordcloud = WordCloud(stopwords=STOPWORDS, 
               background_color="white",
               max_words=300, max_font_size=256,  width=custom_mask_cat.shape[1],
               height=custom_mask_cat.shape[0], mask=transformed_mask_cat, contour_width=3, contour_color='steelblue')

wordcloud.generate(catname)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top Cat names')

    
```

## Pet's images by AdoptionSpeed

```python
#images = os.listdir('petfinder-adoption-prediction/train_images/')
images = [i.split('-')[0] for i in os.listdir('petfinder-adoption-prediction/train_images/')]

pet_type = {1: 'Dog', 2: 'Cat'}
adpt_spd = {0: 'Same Day', 1: 'Between 1 and 7 days', 2: 'Between 8 and 30 days', 3: 'Between 31 and 90 days', 4: 'No'}

for t in train['Type'].unique():
    train_pet = train.loc[(train['Type'] == t)]
    adp = list(train_pet['AdoptionSpeed'].value_counts().index)[:5]
    t = pet_type[t]
    
    fig = plt.figure(figsize=(25, 4))
    print(f"Top {t}s images by Adoption Speed")
    for i, adp_ct in enumerate(adp):
        # excluding pets without pictures
        train_pet_image = train_pet.loc[(train_pet['AdoptionSpeed'] == adp_ct) & (train_pet['PetID'].isin(images)), 'PetID']
        if len(train_pet_image) > 1:
            pet_id = train_pet_image.values[1]
        else:
            pet_id = train_pet_image.values[0]
        ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])

        pet_image = Image.open("petfinder-adoption-prediction/train_images/" + pet_id + '-1.jpg')
        plt.imshow(pet_image)
        adp_ct = adpt_spd[i]
        ax.set_title(f'Adopted: {adp_ct}')
    plt.show();

```

Now that we have done our analysis let's try to model our train data to find a good model to predict adoption speed of pets in the test dataset.

### Random Forest
Firstly I am using Random forest, by splitting the train dataset and finding accuracy to the known portion of train dataset. To start with I am only trying to predict AdoptionSpeed by Type and Age of the pet.


```python
#X_pet = train.drop(['Name','Description','RescuerID','PetID', 'State','AdoptionSpeed'], axis=1)
X_pet = train.drop(['Name', 'Breed1', 'Breed2', 'Gender', 'Color1', 'Color2', 'Color3', 'MaturitySize', 'FurLength'
                    , 'Vaccinated', 'Dewormed', 'Sterilized', 'Health', 'Quantity', 'Fee', 'State', 'RescuerID', 'VideoAmt'
                    , 'Description', 'PetID', 'PhotoAmt', 'AdoptionSpeed'], axis=1)



X_pet.shape
Y_pet = train['AdoptionSpeed']
Y_pet.shape

Xtrain, Xtest, ytrain, ytest = train_test_split(X_pet, Y_pet, random_state=1)
print(Xtrain.shape)
print(ytrain.shape)
print(Xtest.shape)
print(ytest.shape)


model = LinearRegression()
model.fit(Xtrain, ytrain)


r_sq = model.score(Xtrain, ytrain)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)



y_pred = model.predict(Xtest)
print('predicted response:', y_pred, sep='\n')

# predicting the accuracy score
score=r2_score(ytest,y_pred)
print('r2 socre is ',score)
print('mean_sqrd_error is==',mean_squared_error(ytest,y_pred))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(ytest,y_pred)))

```

```python
X_pet = train.drop(['Name', 'Breed1', 'Breed2', 'Gender', 'Color1', 'Color2', 'Color3', 'MaturitySize', 'FurLength'
                    , 'Vaccinated', 'Dewormed', 'Sterilized', 'Health', 'Quantity', 'Fee', 'State', 'RescuerID', 'VideoAmt'
                    , 'Description', 'PetID', 'PhotoAmt', 'AdoptionSpeed'], axis=1)

#X_pet = train(['Type','Age'], axis=1)

X_pet.shape
Y_pet = train['AdoptionSpeed']
Y_pet.shape

seed = 7
test_size = 0.33

X_train, X_test, y_train, y_test = train_test_split(X_pet, Y_pet, test_size=test_size, random_state=seed)
# fit model no training data

model = XGBClassifier(use_label_encoder =False)
model.fit(X_train, y_train)

# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]

# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

```
